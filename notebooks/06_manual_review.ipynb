```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(
    "../data/db/nifty100.db"
)

sample_companies = [
    "ABB",
    "ADANIPORTS",
    "AXISBANK",
    "HDFCBANK",
    "RELIANCE"
]

for company in sample_companies:

    print("\n" + "="*50)
    print(company)

    df = pd.read_sql_query(
        f"""
        SELECT *
        FROM profitandloss
        WHERE company_id='{company}'
        """,
        conn
    )

    print(df.head())

conn.close()
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(
    "../data/db/nifty100.db"
)

coverage = pd.read_sql_query(
    """
    SELECT
        company_id,
        COUNT(*) as years_available
    FROM profitandloss
    GROUP BY company_id
    ORDER BY years_available
    """,
    conn
)

print(
    coverage.head(20)
)

conn.close()
```

```python
print(
    coverage[
        coverage["years_available"] < 5
    ]
)
```

```python
import sqlite3

conn = sqlite3.connect(
    "../data/db/nifty100.db"
)

cursor = conn.cursor()

tables = [
    "companies",
    "profitandloss",
    "balancesheet",
    "cashflow"
]

for table in tables:

    cursor.execute(
        f"SELECT COUNT(*) FROM {table}"
    )

    print(
        table,
        cursor.fetchone()[0]
    )

conn.close()
```

```python
import sqlite3

conn = sqlite3.connect(
    "../data/db/nifty100.db"
)

conn.execute(
    "PRAGMA foreign_keys = ON"
)

cursor = conn.cursor()

cursor.execute(
    "PRAGMA foreign_key_check"
)

violations = cursor.fetchall()

print(
    "FK Violations:",
    len(violations)
)

conn.close()
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(
    "../data/db/nifty100.db"
)

query = """
SELECT DISTINCT company_id
FROM cashflow
WHERE company_id NOT IN (
    SELECT company_id
    FROM companies
)
"""

print(
    pd.read_sql_query(query, conn)
)

conn.close()
```
